```python
# ***
# Setup
# ***
import pandas as pd
import pymysql
import matplotlib
import seaborn as sns
import matplotlib.pyplot as plt
import os
import numpy as np
import scipy as sp
from datetime import date, timedelta
%pylab inline
pd.options.mode.chained_assignment = None  # default='warn'
MYSQL_CONF_FILE = '/'.join([os.path.expanduser('~'), '.my.cnf'])
DEFAULT_DB = 'handybook'

def open_mysql_conn():
    conn = pymysql.connect(read_default_file = MYSQL_CONF_FILE)
    conn.select_db(DEFAULT_DB)
    return conn

conn = open_mysql_conn()
```

```python
MIN_VARIANT_ID = 30
MAX_VARIANT_ID = 33
```

```python
# claim rate per variant
# 
# As of 6/18/2015
# total claimed by booking = 4428
# total claimed by recommendation = 4517
# Since booking is almost equal to recommendations, we'll ignore that slight error in the SQL
sql = """
SELECT
    IF(mv.variant_id IS NULL, "No Variant", mv.variant_id) AS variant_id,
    IF(v.description IS NULL, "No Variant", v.description) AS variant_name,
    COUNT(DISTINCT rj.booking_id) as total_booking,
    SUM(if(rj.matched_at IS NOT NULL, 1, 0)) as total_claimed_by_recommendation,   
    SUM(if(rj.matched_at IS NOT NULL, 1, 0)) / COUNT(DISTINCT rj.booking_id) * 100 AS claim_rate   # total_claimed_by_recommendation / total_booking * 100 
FROM
    recommended_jobs rj
    LEFT JOIN model_variants mv ON 
        mv.model_type = 'User' 
        AND mv.model_id = rj.provider_id 
        AND mv.variant_id BETWEEN %s AND %s
    LEFT JOIN variants v ON mv.variant_id = v.id
GROUP BY
    mv.variant_id DESC
""" % (MIN_VARIANT_ID, MAX_VARIANT_ID)

claim_rate = pd.read_sql(sql, conn)
claim_rate
```

```python
# time to claim by variants

sql = """
SELECT
    IF(c.variant_name IS NULL, 'No variant', c.variant_name) AS variant_name,
    SUM(c.total_claimed) as total_claimed,
    AVG(avg_ttc_per_provider_by_mins) as avg_ttc_by_mins
FROM
  (SELECT
     v.description as variant_name,
     COUNT(*) as total_claimed,
     SUM(TIMESTAMPDIFF(MINUTE, rj.created_at, rj.matched_at)) / COUNT(*) as avg_ttc_per_provider_by_mins
   FROM 
     recommended_jobs rj
     LEFT JOIN model_variants mv ON mv.model_type = 'User' AND mv.model_id = rj.provider_id AND mv.variant_id BETWEEN %s AND %s
     LEFT JOIN variants v ON mv.variant_id = v.id
   WHERE
     rj.matched_at IS NOT NULL
   GROUP BY 
     rj.provider_id
   ) as c
GROUP BY 
    c.variant_name
""" % (MIN_VARIANT_ID, MAX_VARIANT_ID)

ttc = pd.read_sql(sql, conn)
ttc
```

```python
# completion rate for recommended jobs
sql = """
SELECT
	SUM(if(rj.matched_at IS NOT NULL, 1, 0)) as total_claimed,
	SUM(if(b.id IS NOT NULL, 1, 0)) as total_completed,
	SUM(if(b.id IS NULL, 1, 0)) as total_not_completed,
	SUM(if(b.id IS NOT NULL, 1, 0)) / SUM(if(rj.matched_at IS NOT NULL, 1, 0)) * 100 AS completed_rate    # total_completed / total_claimed * 100
FROM 
	recommended_jobs rj
	LEFT JOIN booking b ON 
		rj.booking_id = b.id 
		AND rj.provider_id = b.provider_id
		AND b.when_cancelled IS NULL
		AND b.confirmed = 1
		AND b.date_start < NOW()
WHERE 
	rj.matched_at IS NOT NULL
"""

completed_rate = pd.read_sql(sql, conn)
completed_rate
```

```python
# completion rate for recommended jobs (by regions)
sql = """
SELECT
	r.region,
	SUM(if(rj.matched_at IS NOT NULL, 1, 0)) as claimed,
	SUM(if(b.id IS NOT NULL, 1, 0)) as completed,
	SUM(if(b.id IS NULL, 1, 0)) as not_completed,
	SUM(if(b.id IS NOT NULL, 1, 0)) / SUM(if(rj.matched_at IS NOT NULL, 1, 0)) * 100 AS completed_rate    # completed / claimed * 100
FROM 
	recommended_jobs rj
	LEFT JOIN booking b ON 
		rj.booking_id = b.id 
		AND rj.provider_id = b.provider_id
		AND b.when_cancelled IS NULL
		AND b.confirmed = 1
		AND b.date_start < NOW()
	LEFT JOIN booking b2 ON   -- To get region
		rj.booking_id = b2.id 
	LEFT JOIN l_region r ON 
		r.region_id = b2.region_id
WHERE 
	rj.matched_at IS NOT NULL
GROUP BY 
	r.region_id
"""

completed_rates_by_regions = pd.read_sql(sql, conn)
completed_rates_by_regions
```

```python
# completion rate for regular jobs (by regions)
sql = """
SELECT
	b.id,
	r.region,
	SUM(IF(cb.id IS NULL, 0, 1)) AS completed,
	SUM(IF(cb.id IS NULL, 1, 0)) AS not_completed,
	SUM(IF(cb.id IS NULL, 0, 1)) / COUNT(*) * 100 AS completed_rate    # completed / total bookings
FROM 
	booking b
	JOIN l_region r ON 
		r.region_id = b.region_id
	LEFT JOIN (SELECT
        	b2.id,
        	r2.region
        FROM 
        	booking b2
        	JOIN l_region r2 ON 
        		r2.region_id = b2.region_id
        WHERE 
        	b2.when_cancelled IS NULL
        	AND b2.confirmed = 1
        	AND b2.date_start >= "2015-05-14 20:19:37" 
        	AND b2.date_start < NOW()
     ) cb ON cb.id = b.id   #completed_bookings
WHERE 
	b.date_start >= "2015-05-14 20:19:37"   # recommended job started
	AND b.date_start < NOW()
GROUP BY 
	r.region_id
"""

regular_completed_rates_by_regions = pd.read_sql(sql, conn)
regular_completed_rates_by_regions
```
